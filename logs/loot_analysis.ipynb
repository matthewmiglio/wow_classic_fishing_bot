```python
import matplotlib.pyplot as plt
from collections import Counter

# Read the file
with open('loot_log.txt', 'r') as file:
    lines = file.readlines()

# Extract loot items
loot_items = [line.strip().split(' ', 1)[1] for line in lines]

loot_items = [l for l in loot_items if 'unknown' not in l]

# Count the occurrences of each loot item
loot_counts = Counter(loot_items)

# Sort loot items by count
sorted_loot_counts = dict(sorted(loot_counts.items(), key=lambda item: item[1], reverse=True))

# Create the bar graph
plt.figure(figsize=(20, 16))
plt.bar(sorted_loot_counts.keys(), sorted_loot_counts.values(), color='skyblue')
plt.xlabel('Loot Item')
plt.ylabel('Count')
plt.title('Count of Each Loot Item')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the graph
plt.show()
```
